```python
# Loading dataset.
df = pd.read_csv('PIP-Nov19.csv', low_memory=False)

df.head(20)
```

```python
# Replacing column names.
df.columns = [
    "Month",
    "Region",
    "Disabling condition",
    "Mobility award type",
    "Male",
    "Extra1",
    "Female",
    "Extra2",
    "Extra3"
]

# Dropping unnecessary columns.
df = df.drop(
    columns=[
        "Extra1", 
        "Extra2", 
        "Extra3"
    ], errors="ignore"
)

# Dropping the first row.
df = df.drop(index=[0]).reset_index(drop=True)
```

```python
# Looking at unique values in the columns.
for col in df.columns.to_list():
    print(f"Unique values of {col} :", df[col].unique())
```

```python
df.columns.to_list()[0:3]
```

```python
# Filling missing values in the 'Month', 
# 'Region', 'Disabling condition'.
for col in enumerate(df.columns.to_list()[0:3]):
    df[col[1]] = df[col[1]].ffill()

df.tail()
```

```python
def clean_month_column(df):
    # Removing NaN values.
    df = df.dropna(subset=['Month']) if df['Month'].isnull().any() else df
    
    # Extracting year from the first 4 characters.
    df['Year'] = df['Month'].str[:4].astype(int)
    
    # Extracting month name, handling the specific format.
    df['Month_name'] = df['Month'].str.extract(r'\((.*?)-\d{2}\)')[0]
    
    # Mapping month abbreviations to numbers.
    month_map = {
        'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
        'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
    }
    df['Month_num'] = df['Month_name'].map(month_map)
    
    # Creating a proper datetime column.
    df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month_num'].astype(str) + '-01')
    
    # Dropping the original messy column.
    df = df.drop(columns=['Month'])
    
    return df
```

```python
df = clean_month_column(df=df)
```

```python
df.head()
```

```python
# Melting the columns.
melted_df = pd.melt(
    df, 
    id_vars=[
        'Year', 
        'Month_name',
        'Month_num',
        'Date',  
        'Region', 
        'Disabling condition', 
        'Mobility award type'
    ], value_vars=[
        'Male', 
        'Female'
    ],var_name='Gender', 
    value_name='Population count'
)

melted_df.head()
```

```python
def clean_population_column(df):
    # Replacing '..' with NaN.
    df['Population count'] = df['Population count'].replace('..', np.nan)
    
    # Converting to numeric.
    df['Population count'] = pd.to_numeric(df['Population count'], errors='coerce')
    
    # Removing rows with NaN.
    df = df.dropna(subset=['Population count'])
    
    return df
```

```python
df1 = clean_population_column(df=melted_df).reset_index(drop=True)
```

```python
df1.head()
```

```python
# Visualising Distribution of Mobility Award Types.
plt.figure(figsize=(12, 6))
award_type_counts = df1['Mobility award type'].value_counts()
plt.pie(award_type_counts, labels=award_type_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Mobility Award Types', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Visualising Disability Conditions by Region (Heatmap).
plt.figure(figsize=(16, 10))
conditions_by_region = df1.groupby([
    'Region', 
    'Disabling condition'
])['Population count'].sum().unstack()
sns.heatmap(conditions_by_region, annot=False, cmap='YlGnBu', fmt='.0f')
plt.title('Population Count of Disabling Conditions by Region', fontsize=16)
plt.xlabel('Disabling Condition', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Gender Distribution Across Mobility Award Types.
plt.figure(figsize=(12, 6))
gender_award_type = pd.crosstab(df1['Mobility award type'], df1['Gender'], values=df1['Population count'], aggfunc='sum')
gender_award_type.plot(kind='bar', stacked=True)
plt.title('Population Count by Gender and Mobility Award Type', fontsize=16)
plt.xlabel('Mobility Award Type', fontsize=12)
plt.ylabel('Population Count', fontsize=12)
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
def create_disability_insights(df):
    # Ensuring Population count is numeric.
    df['Population count'] = pd.to_numeric(
        df['Population count'], errors='coerce'
    )
    
    plt.close('all')
    plt.rcParams['figure.figsize'] = (15, 8)
    plt.rcParams['font.size'] = 10
    
    # 1. Distribution of Mobility Award Types.
    plt.figure(figsize=(12, 6))
    award_type_counts = df['Mobility award type'].value_counts()
    plt.pie(award_type_counts, labels=award_type_counts.index, autopct='%1.1f%%')
    plt.title('Distribution of Mobility Award Types', fontsize=16)
    plt.tight_layout()
    plt.show()

    # 2. Disability Conditions by Region (Heatmap).
    plt.figure(figsize=(16, 10))
    conditions_by_region = df.groupby([
        'Region', 'Disabling condition'
    ])['Population count'].sum().unstack()
    sns.heatmap(conditions_by_region, annot=False, cmap='YlGnBu', fmt='.0f')
    plt.title('Population Count of Disabling Conditions by Region', fontsize=16)
    plt.xlabel('Disabling Condition', fontsize=12)
    plt.ylabel('Region', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # 3. Gender Distribution Across Mobility Award Types.
    plt.figure(figsize=(12, 6))
    gender_award_type = pd.crosstab(
        df['Mobility award type'],
        df['Gender'], 
        values=df['Population count'], 
        aggfunc='sum'
    )
    gender_award_type.plot(kind='bar', stacked=True)
    plt.title('Population Count by Gender and Mobility Award Type', fontsize=16)
    plt.xlabel('Mobility Award Type', fontsize=12)
    plt.ylabel('Population Count', fontsize=12)
    plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # 4. Population Count by Disabling Condition.
    plt.figure(figsize=(14, 7))
    population_by_condition = df.groupby(
        'Disabling condition'
    )['Population count'].sum().sort_values(ascending=False)
    population_by_condition.plot(kind='bar')
    plt.title('Total Population Count by Disabling Condition', fontsize=16)
    plt.xlabel('Disabling Condition', fontsize=12)
    plt.ylabel('Total Population Count', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # 5. Temporal Trend of Population Count.
    plt.figure(figsize=(15, 7))
    population_by_year = df.groupby([
        'Year', 'Mobility award type'
    ])['Population count'].sum().unstack()
    population_by_year.plot(kind='line', marker='o')
    plt.title('Population Count Trend by Mobility Award Type', fontsize=16)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Population Count', fontsize=12)
    plt.legend(title='Award Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Generatng a comprehensive summary report.
    report = f"""
    Disability Data Insights Report
    -------------------------------
    1. Mobility Award Types Distribution:
    {award_type_counts}

    2. Top Disabling Conditions by Population:
    {population_by_condition.head()}

    3. Gender Distribution by Award Type:
    {gender_award_type}

    4. Regional Analysis of Disabling Conditions:
    {conditions_by_region.sum()}
    """
    
    with open('disability_insights_report.txt', 'w') as f:
        f.write(report)

    return report

# Calling the function to generate insights.
insights_report = create_disability_insights(df1)
print(insights_report)
```

```python
def create_condition_award_type_plot(df):
    # Creating a pivot table of population count.
    condition_award_pivot = df.pivot_table(
        values='Population count', 
        index='Disabling condition', 
        columns='Mobility award type', 
        aggfunc='sum', 
        fill_value=0
    )
    
    # Plotting heatmap.
    plt.figure(figsize=(16, 10))
    sns.heatmap(
        condition_award_pivot, 
        annot=True, 
        cmap='YlGnBu', 
        fmt='.0f', 
        cbar_kws={'label': 'Population Count'}
    )
    plt.title(
        'Population Count: Disabling Conditions vs Mobility Award Types', 
        fontsize=16
    )
    plt.xlabel('Mobility Award Type', fontsize=12)
    plt.ylabel('Disabling Condition', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    # Stacked bar plot.
    plt.figure(figsize=(16, 10))
    condition_award_pivot.plot(kind='bar', stacked=True)
    plt.title(
        'Distribution of Mobility Award Types Across Disabling Conditions', 
        fontsize=16
    )
    plt.xlabel('Disabling Condition', fontsize=12)
    plt.ylabel('Population Count', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(
        title='Mobility Award Type',
        bbox_to_anchor=(1.05, 1),
        loc='upper left'
    )
    plt.tight_layout()
    plt.show()

    return condition_award_pivot

# Calling the function.
condition_award_pivot = create_condition_award_type_plot(df1)
print(condition_award_pivot)
```
